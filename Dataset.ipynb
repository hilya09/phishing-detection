#Combine

```python
# Gabungkan semuanya
combined_all = pd.concat([data6_bad, data2new, data, data3new, data7new, data4new, data5new])
```

```python
combined_all
```

```python
combined_all.duplicated().sum()
```

```python
combined_all.drop(combined_all[combined_all.duplicated() == True].index, axis = 0, inplace = True)
combined_all.reset_index(drop=True)
```

```python
combined_all.isnull().sum()
```

```python
combined_all.duplicated().sum()
```

```python
def polaritas(value):
  if value =='good':
    return 0
  elif value =='bad':
    return 1
  elif value =='phishing':
    return 1
  elif value =='legitimate':
    return 0
  elif value =='yes':
    return 1
  elif value ==None:
    return 0
  elif value == -1:
    return 0
  elif value == 1:
    return 1
```

```python
combined_all['Label'] = combined_all['Label'].apply(polaritas)
```

```python
combined_all['Label'].unique()
```

```python
print('Total Jumlah url:', combined_all.shape[0], 'data\n')
print('terdiri dari (label):')
print('-- [1] Phishing\t\t:', combined_all[combined_all.label == 1].shape[0], 'data')
print('-- [0] Good\t\t:', combined_all[combined_all.label == 0].shape[0], 'data\n')
```

```python
# Konversi data menjadi DataFrame
df = pd.DataFrame(combined_all, columns=['label'])

# Hitung jumlah masing-masing label
label_count = df['label'].value_counts()

# Buat pie chart menggunakan Seaborn
plt.title('Komposisi Jumlah Dataset URLs')
plt.pie(label_count, labels=label_count.index, autopct='%1.1f%%')

# Tampilkan pie chart
plt.axis('equal')
plt.show()
```
